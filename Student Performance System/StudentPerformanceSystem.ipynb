```python
import numpy as np


class StudentAnalytics:
    """
    Student Performance Analytics System

    Features:
    - Load CSV data
    - Store marks using NumPy
    - Average calculation
    - Topper detection
    - Subject analysis
    - Pass / Fail prediction
    - Rank generation
    - Student search
    - Report export
    """

    def __init__(self):
        """
        Initialize empty containers.
        """

        self.rolls = []
        self.names = []
        self.marks = None

    def is_data_loaded(self):
        """
        Check whether data is loaded.
        """

        if self.marks is None:
            print("\nPlease load data first.\n")
            return False

        return True

    def load_data(self, filename):
        """
        Read CSV and load data.
        """

        try:

            with open(filename, "r") as file:

                rows = file.readlines()[1:]

                rolls = []
                names = []
                marks = []

                for row in rows:

                    data = row.strip().split(",")

                    rolls.append(
                        int(data[0])
                    )

                    names.append(
                        data[1]
                    )

                    marks.append([
                        int(data[2]),
                        int(data[3]),
                        int(data[4])
                    ])

                self.rolls = rolls
                self.names = names
                self.marks = np.array(marks)

                print(
                    "\nData Loaded Successfully\n"
                )

        except FileNotFoundError:

            print(
                "\nCSV file not found.\n"
            )

        except Exception as error:

            print(
                "\nError:",
                error
            )

    def show_students(self):
        """
        Display all students.
        """

        if not self.is_data_loaded():
            return

        print(
            "\n===== STUDENT RECORDS =====\n"
        )

        for i in range(
                len(
                    self.names
                )):

            print(
                f"Roll : {self.rolls[i]}"
            )

            print(
                f"Name : {self.names[i]}"
            )

            print(
                f"Marks: {self.marks[i].tolist()}"
            )

            print(
                "-" * 30
            )

    def average(self):
        """
        Calculate average.
        """

        if not self.is_data_loaded():
            return

        avg = np.mean(
            self.marks,
            axis=1
        )

        print(
            "\n===== AVERAGE =====\n"
        )

        for n, a in zip(
                self.names,
                avg):

            print(
                f"{n} --> {round(a,2)}"
            )

    def topper(self):
        """
        Find topper.
        """

        if not self.is_data_loaded():
            return

        total = np.sum(
            self.marks,
            axis=1
        )

        index = np.argmax(
            total
        )

        print(
            "\n===== TOPPER ====="
        )

        print(
            f"\nName : {self.names[index]}"
        )

        print(
            f"Total: {total[index]}"
        )

    def subject_analysis(self):
        """
        Subject statistics.
        """

        if not self.is_data_loaded():
            return

        subjects = [
            "Math",
            "Physics",
            "Chemistry"
        ]

        print(
            "\n===== SUBJECT ANALYSIS =====\n"
        )

        for i in range(3):

            print(
                subjects[i]
            )

            print(
                "Average:",
                round(
                    np.mean(
                        self.marks[:, i]
                    ),
                    2
                )
            )

            print(
                "Highest:",
                np.max(
                    self.marks[:, i]
                )
            )

            print(
                "Lowest:",
                np.min(
                    self.marks[:, i]
                )
            )

            print(
                "-" * 30
            )

    def pass_fail(self):
        """
        Pass / Fail.
        """

        if not self.is_data_loaded():
            return

        avg = np.mean(
            self.marks,
            axis=1
        )

        print(
            "\n===== RESULT =====\n"
        )

        for i in range(
                len(
                    avg
                )):

            status = (
                "PASS"
                if avg[i] >= 40
                else "FAIL"
            )

            print(
                self.names[i],
                status
            )

    def rank_list(self):
        """
        Generate rank list.
        """

        if not self.is_data_loaded():
            return

        total = np.sum(
            self.marks,
            axis=1
        )

        order = np.argsort(
            total
        )[::-1]

        print(
            "\n===== RANK LIST =====\n"
        )

        rank = 1

        for i in order:

            print(
                rank,
                self.names[i],
                total[i]
            )

            rank += 1

    def search_student(self):
        """
        Search by roll number.
        """

        if not self.is_data_loaded():
            return

        try:

            roll = int(
                input(
                    "\nEnter Roll Number: "
                )
            )

            found = False

            for i in range(
                    len(
                        self.rolls
                    )):

                if self.rolls[i] == roll:

                    found = True

                    print(
                        "\n===== STUDENT FOUND =====\n"
                    )

                    print(
                        f"Roll : {self.rolls[i]}"
                    )

                    print(
                        f"Name : {self.names[i]}"
                    )

                    print(
                        f"Marks: {self.marks[i].tolist()}"
                    )

                    print(
                        f"Average: {round(np.mean(self.marks[i]),2)}"
                    )

                    break

            if not found:

                print(
                    "\nStudent not found.\n"
                )

        except ValueError:

            print(
                "\nEnter numbers only.\n"
            )

    def export_report(self):
        """
        Generate and display report.
        """

        if not self.is_data_loaded():
            return

        try:

            report = ""

            total = np.sum(
                self.marks,
                axis=1
            )

            avg = np.mean(
                self.marks,
                axis=1
            )

            report += (
                "\n===== STUDENT REPORT =====\n\n"
            )

            for i in range(
                    len(
                        self.names
                    )):

                report += f"""
Roll Number : {self.rolls[i]}
Name        : {self.names[i]}
Marks       : {self.marks[i].tolist()}
Total       : {total[i]}
Average     : {round(avg[i],2)}

----------------------------------
"""

            with open(
                    "report.txt",
                    "w"
            ) as file:

                file.write(
                    report
                )

            print(
                report
            )

            print(
                "\nReport Generated Successfully\n"
            )

        except Exception as error:

            print(
                "\nExport Error:",
                error
            )


def menu():
    """
    Application controller.
    """

    app = StudentAnalytics()

    while True:

        print("""
==============================

1 Load Data
2 Show Students
3 Average
4 Topper
5 Subject Analysis
6 Pass / Fail
7 Rank List
8 Search Student
9 Export Report
0 Exit

==============================
""")

        choice = input(
            "Choose: "
        )

        if choice == "1":

            app.load_data(
                "data.csv"
            )

        elif choice == "2":

            app.show_students()

        elif choice == "3":

            app.average()

        elif choice == "4":

            app.topper()

        elif choice == "5":

            app.subject_analysis()

        elif choice == "6":

            app.pass_fail()

        elif choice == "7":

            app.rank_list()

        elif choice == "8":

            app.search_student()

        elif choice == "9":

            app.export_report()

        elif choice == "0":

            print(
                "\nProgram Closed"
            )

            break

        else:

            print(
                "\nInvalid Choice\n"
            )


if __name__ == "__main__":
    menu()

```
